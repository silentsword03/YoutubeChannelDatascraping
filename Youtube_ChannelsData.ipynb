```python
import requests

def fetch_html(url):
    response = requests.get(url)
    html = response.text
    return html
```

```python
import pandas as pd
from bs4 import BeautifulSoup

links = DF_Final['Links']

data = []
columns = ["Links", "Title", "Created Date", "Country", "Average Likes", "Est. Monthly Income", "Est. Yearly Income"]

for link in links:
    html = fetch_html(link)
    soup = BeautifulSoup(html, 'html.parser')
    
    row_data = [link]
    for card in soup.select('.card'):
        card_title_element = card.select_one('.card-title')
        card_value_element = card.select_one('.badge-primary') or card.select_one('.badge-success')
        
        if card_title_element and card_value_element:
            card_title = card_title_element.text.strip()
            card_value = card_value_element.text.strip()
            row_data.append(card_value)
    
    # Check if row_data has the expected length of columns
    if len(row_data) == len(columns):
        data.append(row_data)

# Identify missing links
missing_links = set(links) - set(df['Links'])

# Append missing links with missing values
for link in missing_links:
    row_data = [link]
    for _ in range(len(columns) - 1):
        row_data.append(None)  # or row_data.append('')
    
    data.append(row_data)

df = pd.DataFrame(data, columns=columns)
print(df)
```

```python
merged_df['Subscribers'] = pd.to_numeric(merged_df['Subscribers'].astype(str).str.replace('M', '')) * 1e6
merged_df['Videos'] = pd.to_numeric(merged_df['Videos'])
```

```python
merged_df['Created Date'] = pd.to_datetime(merged_df['Created Date'])
```

```python
merged_df['Average Likes'] = pd.to_numeric(merged_df['Average Likes'].str.replace('K', ''))
merged_df['Average Comments'] = pd.to_numeric(merged_df['Average Comments'].str.replace('K', ''))
```

```python
merged_df['Views'] = merged_df['Views'].astype(str).apply(lambda x: float(x[:-1]) * 1e9 if x.endswith('B') else float(x[:-1]) * 1e6)
```

```python
merged_df[['Est. Monthly Income (Low)', 'Est. Monthly Income (High)']] = merged_df['Est. Monthly Income'].str.extract(r'\$(\d+\.?\d*)[KM] - \$(\d+\.?\d*)[KM]').astype(float) * 1e3
merged_df[['Est. Yearly Income (Low)', 'Est. Yearly Income (High)']] = merged_df['Est. Yearly Income'].str.extract(r'\$(\d+\.?\d*)[KM] - \$(\d+\.?\d*)[KM]').astype(float) * 1e6
```
